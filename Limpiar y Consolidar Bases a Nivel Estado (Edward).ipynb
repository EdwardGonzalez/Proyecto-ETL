```python
#Importar csv

#modules
import os
import csv
import pandas as pd

#Import CSV
csvpath = os.path.join("Raw Data", "AlcoholismoDrogasEntidadFederativa.csv")
# Import the Bancos.csv file as a DataFrame
DrogasDB = pd.read_csv(csvpath, encoding="utf-8")

#Import CSV
csvpath = os.path.join("Raw Data", "CatalogoEntidad.csv")
# Import the Bancos.csv file as a DataFrame
CatalogoEntidades = pd.read_csv(csvpath, encoding="utf-8")

#Import CSV
csvpath = os.path.join("Raw Data", "DesempleoPorEntidadFederativa.csv")
# Import the Bancos.csv file as a DataFrame
DesempleoDB = pd.read_csv(csvpath, encoding="utf-8")

#Import CSV
csvpath = os.path.join("Raw Data", "ParqueVehicular.csv")
# Import the Bancos.csv file as a DataFrame
ParqueVehicularDB = pd.read_csv(csvpath, encoding="utf-8")

#Import CSV
csvpath = os.path.join("Raw Data", "PoblacionEntidadFederativa.csv")
# Import the Bancos.csv file as a DataFrame
PoblacionDB = pd.read_csv(csvpath, encoding="utf-8")

#Import CSV
csvpath = os.path.join("Raw Data", "UsuariosConAccesoInternetEntidadFederativa.csv")
# Import the Bancos.csv file as a DataFrame
UsuariosInternetDB = pd.read_csv(csvpath, encoding="utf-8")
```

```python
#Ponerlos ID del catalogo en cada Base

#Drogas, sacar el promedio por estado
DrogasDB = DrogasDB.groupby("Estado").mean()
#Desempleo, Eliminar columnas
DesempleoDB = DesempleoDB.drop(columns=['q2','q3','q4','q1'])
#ParqueVeihicular, elimnar columna y cambiar titulo
ParqueVehicularDB = ParqueVehicularDB.rename(columns={"ENTIDAD FEDERATIVA": "Estado"})
#Poblacion, Eliminar comas cambiar type y sumar
PoblacionDB['Personas'] = PoblacionDB['Personas'].str.replace(',','')
PoblacionDB['Personas'] = pd.to_numeric(PoblacionDB['Personas'])
PoblacionDB=PoblacionDB.groupby('Estado').sum()
#Usuarios Internet
UsuariosInternetDB = UsuariosInternetDB.rename(columns={"Entidad Federativa": "Estado"})
```

```python
#Colocar ID de Estado
DrogasDB = DrogasDB.merge(CatalogoEntidades, on='Estado',how='left')
DesempleoDB = DesempleoDB.merge(CatalogoEntidades, on='Estado',how='left')
ParqueVehicularDB = ParqueVehicularDB.merge(CatalogoEntidades, on='Estado',how='left')
PoblacionDB = PoblacionDB.merge(CatalogoEntidades, on='Estado',how='left')
UsuariosInternetDB = UsuariosInternetDB.merge(CatalogoEntidades, on='Estado',how='left')

```

```python
#Crear MainDB, en base al catalogo
MainDB = pd.DataFrame ({
    "ID Estado":[]
})

MainDB['ID Estado'] = CatalogoEntidades['ID Estado'].unique()

```

```python
#Juntar Base, y renombar para SQL
MainDB = MainDB.merge(DrogasDB, on='ID Estado', how='left')
MainDB = MainDB.drop(columns=['Estado'])

MainDB = MainDB.merge(DesempleoDB, on='ID Estado', how='left')
MainDB = MainDB.drop(columns=['Estado', 'Año'])
MainDB = MainDB.rename(columns={"Total": "%_Desempleados"})

MainDB = MainDB.merge(ParqueVehicularDB, on='ID Estado', how='left')
MainDB = MainDB.drop(columns=['Estado'])
MainDB = MainDB.rename(columns={"TOTAL": "Total_Vehiculos",
                                "AUTOMÓVILES": "Total_de_Automoviles",
                               "AUTOMÓVILES": "Total_de_Automoviles",
                               "CAMIONES PARA PASAJEROS": "Total_Camiones_Pasajeros",
                               "CAMIONES Y CAMIONETAS PARA CARGA": "Total_Camiones_Carga",
                               "MOTOCICLETAS": "Total_Motocicletas"
                               })

MainDB = MainDB.merge(PoblacionDB, on='ID Estado', how='left')
MainDB = MainDB.drop(columns=['Estado'])
MainDB = MainDB.rename(columns={"Personas": "Poblacion"})

MainDB = MainDB.merge(UsuariosInternetDB, on='ID Estado', how='left')
MainDB = MainDB.drop(columns=['Estado'])

MainDB = MainDB.rename(columns={"ID Estado": "ID_ENTIDAD",
                                "Consumo Litros de Alcohol": "Consumo_Alcohol_Litros",
                               "Consumo de Mariguana": "Consumo_Mariguana",
                               "Consumo de Cocaina": "Consumo_Cocaina",
                               "CAMIONES Y CAMIONETAS PARA CARGA": "Total_Camiones_Carga",
                               "MOTOCICLETAS": "Total_Motocicletas",
                                "Usuarios con acceso a internet":"Usuarios_Acceso_Internet"
                               })


MainDB
```

```python
#imprimir Archivo
MainDB.to_csv(r'MainDBnivelEstado.csv')
```
